```python
import pandas as pd
df = pd.read_csv("store.csv")
df.info()
df.head()
df.columns
```

```python
df.columns = (
    df.columns.str.strip().str.lower()
.str.replace(" " , "_")
.str.replace("é" , "e")
.str.replace("à" , "a")
.str.replace("è" , "e")
) 
print(df.columns)
```

```python
df = df.dropna(subset=["order_date", "sales"])
df = df.drop_duplicates()
```

```python
df['order_date'] = pd.to_datetime(df["order_date"], errors='coerce')
df['ship_date'] = pd.to_datetime(df["ship_date"], errors='coerce')
```

```python
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
df["quartier"] = df["order_date"].dt.to_period('Q').astype(str)
print(df[["order_date", "year", "month", "quartier"]].head())
```

```python
df["delivery"] = (df['ship_date'] - df["order_date"]).dt.days
```

```python
cost = df["sales"] * 0.6
marge = df["sales"] - cost
df["tp_prft"] = marge / df["sales"] * 100
```

```python
c_a = df["sales"].sum()
```

```python
mensuel = df.groupby("month")["sales"].sum()
croissance = mensuel.pct_change() * 100 
print(croissance)
```

```python
df["delivery_late"] = df["delivery"] > 3
```

```python
top_clients = df.groupby("customer_name")["sales"].sum()
top_clients = top_clients.sort_values(ascending=False)
df["top_clients_sales"] = df["customer_name"].map(top_clients)
print(top_clients.head(10)) 
```

```python
top_produits = (
    df.groupby("product_name")["sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

sales_region = (
    df.groupby("region")["sales"]
    .sum()
    .sort_values(ascending=False)
)
sales_categories =(
    df.groupby('category')["sales"]
    .sum()
    .sort_values(ascending=False)
    
) 
```

```python
df.to_csv("projet.csv", index=False)
```
